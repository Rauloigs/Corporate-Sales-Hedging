# Metodología Cuantitativa de Selección de Activos: Arquitectura Core-Satellite y Cobertura Sectorial (GICS)

El presente modelo cuantitativo tiene como objetivo diseñar un portafolio de inversión para los excedentes de liquidez de la tesorería corporativa. A diferencia de un fondo de inversión tradicional, el mandato de una tesorería no es maximizar el retorno absoluto asumiendo riesgos indiscriminados, sino **preservar el valor real del capital, garantizar la eficiencia fiscal (diferimiento) y neutralizar las fricciones del ciclo operativo local**. 

Para lograr esto sin incurrir en sesgos empíricos o heurísticas frágiles (como la selección arbitraria de activos), el algoritmo de selección se fundamenta en una arquitectura institucional **Core-Satellite (Núcleo-Satélite)**, soportada por la Teoría Moderna de Portafolios y el marco analítico de Fama y French.



---

### 1. El Núcleo (Core): Capturando la Prima de Riesgo Global (Fama-French & Bogle)

La base estructural del portafolio estará conformada por el índice de mercado amplio de Estados Unidos, representado por el ETF **IVV (S&P 500 Total)**. 

**Fundamentación Académica:**
La decisión de utilizar un vehículo pasivo de mercado total como núcleo se sustenta en la **Hipótesis de Mercados Eficientes (EMH)** y el **Modelo de Tres Factores de Fama y French (1993)**. Eugene Fama y Kenneth French demostraron matemáticamente que el riesgo de mercado (el factor *Market Beta* o Prima de Riesgo de Capital - ERP) explica la inmensa mayoría de los retornos de un portafolio diversificado a largo plazo. 



Intentar vencer al mercado mediante la selección activa de acciones individuales (*Stock-Picking*) dentro de la tesorería corporativa introduce un riesgo idiosincrático innecesario y costos de fricción que destruyen valor. Como lo documentan Fahling, Steurer & Sauer (2019) y la filosofía fundacional de John Bogle, la captura pasiva del índice (IVV) garantiza la obtención del rendimiento histórico del capital productivo global con la máxima liquidez, la menor varianza posible y costos de intermediación cercanos a cero (TER de 0.03%). 

En resumen: El IVV cumple la función de ancla patrimonial para el crecimiento a largo plazo.

### 3. El Satélite (Hedge Tilt): Cobertura Operativa de la Tesorería

La literatura financiera sobre *Corporate Hedging* (Gupta, 2017) establece que la tesorería debe neutralizar las fricciones específicas de su ciclo operativo. El portafolio no necesita 15 activos descorrelacionados al azar; necesita **un ancla anticíclica**.

* **El Mecanismo de Selección:** El modelo computacional escaneará 10 vectores sectoriales del GICS (El  ETF XLC no esta dentro del rango 2018 - 2025)frente a los retornos históricos de las ventas de la empresa (`ventas_log_ret`). 
* **La Inclinación de Cobertura (*Hedge Tilt*):** El sector económico que presente el coeficiente de correlación de Pearson (\(\rho\)) más bajo y/o negativo será designado matemáticamente como el "Satélite". 
* **La Función:** Este activo satélite recibirá un sobrepeso (*overweight*) en la asignación final. Su única misión económica es actuar como un amortiguador: cuando el ciclo de capital local de la empresa se contraiga, este sector global específico debería expandirse o mantener su valor, protegiendo la liquidez general del balance.

### Conclusión de Robustez
Esta arquitectura garantiza la universalidad y replicabilidad del estudio. El método está blindado contra la subjetividad y se adapta dinámicamente al ADN financiero de cualquier organización.

#### Ingeniería de Datos: Transformación de Divisas (Método Log-Aditivo)

**El Problema Metodológico:**
Para evaluar correctamente el desempeño de los activos globales (cotizados en USD) desde la perspectiva de una tesorería corporativa en México, es imperativo analizar sus rendimientos en pesos mexicanos (MXN) para capturar el efecto del riesgo cambiario. 

El enfoque tradicional consiste en calcular el precio en pesos multiplicando el precio del ETF por el tipo de cambio, y luego aplicar la fórmula de variación con un rezago de un periodo. Sin embargo, computacionalmente esto requiere descartar la primera fila de datos, lo que nos haría perder la observación de Enero 2018 y reduciría el tamaño de nuestra muestra.

**La Solución Cuantitativa:**
Para preservar la integridad de la ventana de tiempo ($N=96$ meses), este modelo aprovecha la **propiedad aditiva de los rendimientos logarítmicos** (composición continua). 

Por las leyes matemáticas de los logaritmos, el logaritmo de un producto es igual a la suma de los logaritmos de sus factores: $\ln(A \times B) = \ln(A) + \ln(B)$. 

Si definimos el precio en pesos como el producto del precio en dólares y el tipo de cambio:
$$P_{MXN, t} = P_{USD, t} \times TC_t$$

La fórmula del retorno logarítmico en pesos se expresa como:
$$r_{MXN, t} = \ln\left(\frac{P_{USD, t} \times TC_t}{P_{USD, t-1} \times TC_{t-1}}\right)$$

Aplicando las propiedades de los logaritmos, podemos separar los términos:
$$r_{MXN, t} = \ln\left(\frac{P_{USD, t}}{P_{USD, t-1}}\right) + \ln\left(\frac{TC_t}{TC_{t-1}}\right)$$

Lo que se reduce a una suma algebraica exacta de las tasas de rendimiento individuales:
$$r_{MXN} = r_{ETF(USD)} + r_{USDMXN}$$

**Beneficio para el Modelo:** Al sumar linealmente los retornos logarítmicos ya existentes en nuestra base de datos, evitamos rezagar la serie temporal, manteniendo la muestra completa desde el primer mes y garantizando que el "efecto divisa" se integre sin pérdida de poder estadístico.

```python
etfs_en_usd = ['IVV', 'QQQ', 'XLV', 'XLE', 'XLP', 'XLI', 
               'GLD', 'XLY', 'XLU', 'XLRE', 'EWJ', 'SHV', 'XLK', 'XLF', 'XLB'] 
df_niveles_total, df_retornos_total = finp.convertir_divisa_y_sumar_retornos(
    df_niveles=df_niveles_total,
    df_retornos=df_retornos_total,
    columnas_etfs=etfs_en_usd,
    col_usdmxn_nivel='usdmxn_fix',  
    col_usdmxn_ret='rets_usdmxn'     
)
```

### Correlación

**A. El Principio de Diversificación Matemática (Markowitz, 1952)**
La Teoría Moderna de Portafolios (MPT) establece inequívocamente que el riesgo de un portafolio no es el promedio simple de los riesgos de sus activos, sino que depende de su covarianza. La regla matemática fundamental es que **cualquier coeficiente de correlación de Pearson ($\rho$) estrictamente menor a 1.0 reduce la varianza total del portafolio.** En el contexto de la tesorería corporativa, esto significa que no necesitamos buscar obsesivamente correlaciones altamente negativas; basta con evitar la redundancia ($\rho > 0.8$). Si las ventas locales y el S&P 500 tienen una correlación de 0.15, son "ortogonales" (independientes), lo cual es una diversificación pura y altamente deseable.

**B. Fundamento Económico vs. Minería de Datos (Spurious Correlation)**
La metodología de esta investigación rechaza explícitamente el *Data Mining* (correlacionar miles de activos indiscriminadamente para encontrar el decil más bajo). En econometría financiera, buscar correlaciones sin un raciocinio económico subyacente genera **correlaciones espurias** que inevitablemente colapsan fuera de la muestra (*out-of-sample*). Por lo tanto, los activos candidatos (ETFs Globales y USDMXN) se seleccionan primero por su **lógica económica** (exposición a la economía global vs. local; cobertura cambiaria natural para una importadora) y la matriz de correlación se utiliza únicamente como herramienta de *validación* estadística, no de *descubrimiento*.

**C. La Definición de Cobertura Corporativa (Corporate Hedging)**
Apoyándonos en Gupta (2017), el objetivo de la tesorería no es especular, sino estabilizar el flujo de caja. Los activos que presenten una correlación $\rho \le 0$ con las ventas (como teóricamente debería ser el USD/MXN para una importadora), actúan como una cobertura natural (*Natural Hedge*). Cuando el negocio operativo enfrenta fricciones macroeconómicas locales, el portafolio financiero mitiga la pérdida de valor del balance general.

## Desempeño Ajustado al Riesgo: Retornos Anualizados y Eficiencia de Capital

La validación fiduciaria de una estrategia de tesorería exige trascender la medición del valor absoluto final para auditar el costo de oportunidad del capital en el tiempo, medido a través de la Tasa Anual de Crecimiento Compuesto (CAGR), y la compensación financiera por la volatilidad asumida, evaluada mediante el Ratio de Sharpe.

El recálculo de los retornos bajo los cuatro escenarios fiscales y operativos arroja datos concluyentes sobre la ineficiencia estructural de los vehículos tradicionales y el mandato del horizonte de inversión:

**1. El Costo de Oportunidad y el Riesgo de Reinversión (Status Quo)**
El fondo de renta fija (CETES) fungió como la tasa libre de riesgo del modelo corporativo, entregando un CAGR del **5.34%**. Es imperativo notar que esta alternativa no está exenta de fricción; registró una volatilidad anualizada del **2.20%** derivada del riesgo de reinversión frente a los ciclos de política monetaria del banco central durante los 96 meses analizados. Si bien garantiza estabilidad de principal, su crecimiento compuesto restringe severamente la maximización patrimonial de la caja.

**2. Destrucción de Valor en el Vehículo Bancarizado (Sharpe Negativo)**
La auditoría expone la inviabilidad financiera de la renta variable tradicional (Portafolio Fantasma). Al someter la estrategia a tributación anual y comisiones de gestión, su CAGR colapsó a un deficiente **3.06%**. Este rendimiento inferior a la tasa libre de riesgo, combinado con una exposición a volatilidad de mercado (14.96%), resultó en un Ratio de Sharpe destructivo de **-0.15**. Matemáticamente, la estructura bancaria obligó a la tesorería a asumir un riesgo de mercado considerable sin otorgar prima de riesgo alguna, configurando un escenario de destrucción de valor corporativo.

**3. La Prima de Riesgo y el Mandato del Horizonte Temporal (GMV)**
La arquitectura GMV Diferida restaura la lógica financiera de la inversión. Al evitar las extracciones anuales, el portafolio estructural (GMV Bruto) logró expandir su CAGR al **7.78%**, superando holgadamente a la renta fija.

## Auditoría de Cobertura Condicional: Prueba de Estrés Operativo y Eficiencia del *Proxy Hedge*

Para validar la eficacia del portafolio óptimo (GMV) como instrumento de *Corporate Hedging*, se ejecutó una **Prueba de Estrés de Rendimiento Condicional**. 

El objetivo de esta métrica es auditar el comportamiento del portafolio de tesorería exclusivamente durante los periodos de contracción operativa de la empresa. Para aislar el ciclo macroeconómico del ruido estacional, se suavizaron las series de tiempo calculando el crecimiento móvil de 12 meses (Trailing Twelve Months - TTM). Utilizar instrumentos de renta variable para cubrir emergencias de capital de trabajo constituiría un error fiduciario; por lo tanto, esta cobertura está diseñada estrictamente para **capital estructural ocioso**.

**2. Rendimiento del Escudo (El Hedge):** Exactamente durante los mismos periodos de colapso comercial, el portafolio GMV (inclinado estratégicamente hacia el sector de Consumo Básico) no solo preservó el capital, sino que generó un rendimiento promedio positivo del **+13.2%**.

Lejos de representar un "gasto de seguro", el instrumento generó riqueza (+13.2%) mientras fungía como escudo protector.

**5. El Doble Beneficio (Interés Compuesto Ininterrumpido):** El valor supremo de esta estrategia radica en la conjunción de la cobertura y la eficiencia fiscal. ## Análisis de Riesgo de Cola (Tail Risk): VaR y Expected Shortfall (CVaR)

Para certificar la idoneidad fiduciaria del portafolio en la gestión de excedentes de tesorería, el análisis de volatilidad tradicional (desviación estándar) es insuficiente, ya que asume una distribución normal de los retornos. Los mercados financieros exhiben asimetría y colas pesadas (*fat tails*); por tanto, el riesgo corporativo debe evaluarse mediante métricas de pérdida extrema en la cola izquierda de la distribución.

Para este análisis se utilizan dos métricas complementarias:
* **Value at Risk (VaR):** Define la frontera del riesgo. Cuantifica la pérdida mensual mínima esperada dentro de un percentil específico de los peores escenarios históricos.
* **Conditional Value at Risk (CVaR / Expected Shortfall):** Mide la profundidad de la crisis. Cuantifica el promedio matemático de las pérdidas que superan el umbral del VaR. Es la métrica definitiva de riesgo sistémico, recomendada por los estándares de Basilea.

Los resultados empíricos (N=96 meses) demuestran la superioridad estructural del Portafolio de Mínima Varianza Global (GMV) frente a la exposición al mercado amplio (S&P 500):

**1. Contención de Riesgo en Escenarios Críticos (Confianza 95%)**
En el 5% de los peores meses históricos, el mercado accionario puro (S&P 500) presentó un VaR de **-9.06%** y una pérdida promedio en esa cola (CVaR) del **-10.40%**. La exposición directa de la tesorería a este vehículo comprometería severamente la liquidez del balance. En contraste, la optimización GMV comprimió la frontera de riesgo a un VaR de **-5.02%**, con un CVaR de **-6.62%**. Esta reducción casi a la mitad del *Expected Shortfall* mensual es el mecanismo directo que explica por qué el Portafolio GMV logró mantener un *Max Drawdown* tan contenido (-19.2%) a lo largo de 8 años: al limitar el daño mensual severo, se previene la destrucción del capital por interés compuesto negativo durante recesiones prolongadas.

**2. Pruebas de Estrés Extremo (Confianza 99%)**
Al someter la estrategia al estándar de rigor fiduciario más estricto (el 1% de los peores escenarios históricos), el Portafolio GMV ratifica su perfil defensivo. Mientras que el riesgo de cola extrema del S&P 500 cae hasta un CVaR del **-11.65%**, el portafolio propuesto encuentra un piso sólido con un CVaR de **-8.46%**. 

**3. Perfil de Distribución (Curtosis)**
Visualmente, la distribución de los retornos del Portafolio GMV presenta características leptocúrticas favorables para la tesorería: una alta concentración de retornos alrededor de su media (mayor predictibilidad) y una drástica reducción en la densidad de la cola izquierda (menor probabilidad de catástrofe financiera) en comparación con el S&P 500.

**Conclusión del Perfil de Riesgo:**
La estrategia GMV exige a la empresa abandonar la certidumbre absoluta de la renta fija (VaR 0%), asumiendo un riesgo de cola esperado (CVaR 95%) del -6.62% mensual. Sin embargo, este es un riesgo controlado, asimétrico y justificado matemáticamente frente a la alternativa de generar decenas de millones en Alpha Latente. Queda comprobado que la arquitectura del portafolio elimina eficientemente los riesgos de cola catastróficos que inhabilitarían la inversión institucional en renta variable.

## 6.2. Normativa Contable y Blindaje del Estado de Resultados (NIF C-2)

La adopción de una estrategia de tesorería invertida en renta variable introduce, por definición, volatilidad de mercado. El principal obstáculo en la gobernanza de empresas familiares o corporativos tradicionales es la aversión al riesgo contable: la intolerancia a que las fluctuaciones temporales de los activos financieros deterioren la percepción de la utilidad operativa.

Si el portafolio GMV se clasifica erróneamente bajo la contabilidad corporativa, la estrategia está destinada al fracaso por pánico gerencial. Bajo las Normas de Información Financiera mexicanas (específicamente la **NIF C-2: Inversión en instrumentos financieros**), es un error fiduciario registrar esta caja estructural como *Instrumentos Financieros Negociables (IFN)*. Hacerlo obligaría a la empresa a registrar el Ajuste a Valor Razonable (*Mark-to-Market*) mensual directamente en el Estado de Resultados Integral (ERI). En un mes de corrección sistémica (ej. marzo 2020), la empresa reportaría una "Pérdida Financiera" masiva, destruyendo la Utilidad Neta del periodo a pesar de que la pérdida no ha sido realizada (vendida) y de que el negocio principal (*core business*) opera con normalidad.



**El Escudo Contable: Clasificación IFCV y el Uso del ORI**
Para alinear la contabilidad con la tesis de inversión a largo plazo (diferimiento de capital estructural), la Política de Inversión (IPS) debe dictaminar desde su origen que el Portafolio GMV se clasifica estrictamente como **Instrumentos Financieros para Cobrar o Vender (IFCV)**. 

Bajo esta clasificación normativa, se logra el blindaje operativo:
1.  **Aislamiento de la Volatilidad:** Las fluctuaciones mensuales de valor razonable (ganancias o pérdidas no realizadas) **no impactan la Utilidad Neta**. En su lugar, se registran directamente en el Capital Contable dentro del rubro de **Otro Resultado Integral (ORI)**.
2.  **Claridad Operativa:** El EBITDA, la Utilidad de Operación y la Utilidad Neta consolidada de la empresa se mantienen puramente atados al desempeño de las ventas industriales. Los accionistas y el Consejo de Administración pueden evaluar la salud del negocio real sin el "ruido blanco" del mercado bursátil.
3.  **Realización Estratégica:** Únicamente en el momento en que el Comité de Inversiones decida ejecutar una venta (liquidación parcial o total), la ganancia o pérdida acumulada se recicla del ORI y se reconoce en el Estado de Resultados, empatando perfectamente el impacto contable con el impacto fiscal (hecho imponible) y el flujo de efectivo real.

**Conclusión Operativa:**
El uso estratégico de la NIF C-2 elimina el riesgo de liquidación por pánico. Al dirigir la volatilidad hacia el Capital Contable, se asegura la tranquilidad psicológica de los accionistas, protegiendo el horizonte temporal requerido para que el *Alpha Latente* y el interés compuesto ejecuten su función matemática.

## Nota Metodológica: Supuestos de Reinversión y Fricción Fiscal (Tax Drag)

Para el desarrollo de la simulación histórica (*backtesting*) y la optimización de varianza, las series de tiempo de los activos subyacentes (IVV y XLP) fueron extraídas utilizando el precio de Cierre Ajustado (*Adjusted Close*). Metodológicamente, este parámetro asume la reinversión bruta total del 100% de los dividendos corporativos distribuidos, sin descontar retenciones impositivas en la fuente.

En una ejecución institucional transfronteriza, donde una entidad moral mexicana opera mediante una cuenta de corretaje en Estados Unidos amparada bajo la forma W-8BEN-E, los dividendos pagados por los ETFs están sujetos a una retención fiscal en origen del 10% bajo el tratado para evitar la doble tributación. 

**Impacto Paramétrico en el Modelo:**
Dado que el rendimiento por dividendos (*Dividend Yield*) histórico promedio oscila en torno al 1.5% para el S&P 500 y 2.5% para el sector de Consumo Básico (XLP), la retención impositiva del 10% sobre estos flujos representa una fricción o *tax drag* estructural estimado entre 15 y 25 puntos base (0.15% - 0.25%) anuales sobre el valor total del portafolio.

Para propósitos de esta investigación, y en congruencia con los estándares académicos de modelado de portafolios, esta fricción se considera un margen de error estadísticamente no significativo que no altera la validez de la Frontera Eficiente ni la superioridad matemática del diferimiento del Impuesto Sobre la Renta (ISR) sobre las ganancias de capital. No obstante, en un entorno de ejecución fiduciaria estricta, este *Slippage* fiscal debe integrarse en las proyecciones de flujo de caja libre de la tesorería.
